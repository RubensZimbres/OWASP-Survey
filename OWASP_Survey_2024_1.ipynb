```python
import statsmodels.api as sm

def norm(x):
    return (x-np.min(x))/(np.max(x)-np.min(x))

X=norm(df24.iloc[:,1:].astype(float))
Y=norm(df24['Usefulness OWASP Top 10 LLM'].astype(float))
X
```

```python
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
results.params
```
